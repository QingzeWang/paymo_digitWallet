```python
import sys
import csv
import time
import pickle

#import warnings
#warnings.filterwarnings("ignore")

batch_dictionary = dict()
start_time = time.time()

with open("../paymo_input/batch_payment.csv",'rU') as f_batch_handle: #rU: open a file with an universal newline mode
    f_batch = csv.reader(f_batch_handle)
    #row_count = sum(1 for row in f_batch)
    #print("Number of rows: %s"%row_count) ## 3934584
    i = 0;

    for line in f_batch:
        
        if (i%5e3 == 0):
            print i, usr1, usr2
        i+=1
#        if i==410609 or i == 410608 or i== 410610:
#            print line
#        if(len(line)!=5 ):
#            try:
#                usr1 = int(line[1])
#                usr2 = int(line[2])
#            except:
#                print i
#                print line

        try:    
            usr1 = int(line[1])
            usr2 = int(line[2])
        except:
            continue
        #if i>5e5: break
        if usr1 not in batch_dictionary:
            batch_dictionary[usr1] = []
            batch_dictionary[usr1].append(usr2)
        else:
            batch_dictionary[usr1] = list(set(batch_dictionary[usr1])) 
            batch_dictionary[usr1].append(usr2)
            
        if usr2 not in batch_dictionary:
            batch_dictionary[usr2] = []
            batch_dictionary[usr2].append(usr1)
        else:
            batch_dictionary[usr2] = list(set(batch_dictionary[usr2])) 
            batch_dictionary[usr2].append(usr1)        
        
f_batch_handle.close()   

print("time elapsed is %s s"%(time.time()-start_time))

with open('paymo_graph.pickle', 'wb') as f_batch_handle:
    pickle.dump(batch_dictionary, f_batch_handle, protocol=pickle.HIGHEST_PROTOCOL)

#dt_batch = pd.read_csv("../paymo_input/batch_payment.csv")
#dt_stream = pd.read_csv("../paymo_input/stream_payment.csv")
```
